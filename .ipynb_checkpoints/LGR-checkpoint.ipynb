```python
import numpy as np
from scipy.signal import lti
import matplotlib.pyplot as plt
```

```python
class tf:
    def __init__(self,num,den):
        self.num=num
        self.den=den
        
    def serie(self,sys):
        '''
        sys=[num,den,ret]
        '''
        num=np.polymul(self.num,sys[0])
        den=np.polymul(self.den,sys[1])
        return tf(num,den)
    
    def feedback(self,sys):
        '''
        sys=[num,den,ret]
        '''
        num=np.polymul(self.num,sys[1])
        den=np.polyadd(np.polymul(self.num,sys[0]),np.polymul(self.den,sys[1]))
        return tf(num,den)
    
    
    def polos(self):
        return lti(self.num,self.den).poles
    
    def ceros(self):
        return lti(self.num,self.den).zeros
    
    def LGR(self,f='.'):
        
        #EXTRAER LOS POLOS Y CEROS DE LA FUNCION DE TRANSFERENCIA DE LAZO DIRECTO
        p=self.polos() 
        z=self.ceros()
        
        real=[]
        imag=[]
        
        k=np.linspace(0.01,50,300)
        for i in k:
            serie=self.serie([[i],[1]])
            fed=serie.feedback([[1],[1]])
            ps=fed.polos()
            real.append(ps.real)
            imag.append(ps.imag)
            
        %matplotlib notebook
        plt.figure(facecolor='grey',edgecolor='black')
        plt.plot([p.real-3,p.real+3],[0]*2,'k',[0]*2,[p.imag+3,-p.imag-3],'k',lw=1)#graficamos los ejes
        plt.plot(p.real,p.imag,c='blue',marker='x',ms=10,ls='')# graficamos los polos del sistema en lazo directo
        plt.plot(z.real,z.imag,c='blue',marker='o',ms=10,ls='')#graficamos los ceros de sistema en lazo directo
        plt.plot(real,imag,f)
        plt.title('LGR')
        plt.xlabel('real')
        plt.ylabel('imag')
        plt.show()

    
```
